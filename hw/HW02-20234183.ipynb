### 가정
- 아래에서 생성된 _goog_ 및 _aapl_ 변수는 2010년부터 2024년까지의 주가에 데이터를 담은 DataFrame이다.

```
### 지우지 마시오.
import pandas as pd
import numpy as np
import yfinance as yf
goog = yf.download('GOOG', start='2010-01-01', end='2024-12-31').droplevel(1,axis=1) # 구글
aapl = yf.download('AAPL', start='2010-01-01', end='2024-12-31').droplevel(1,axis=1) # 애플
aapl.head()
```

### 문제1
두 변수 *goog*, *aapl* 에 대하여 아래 수식으로 정의되는 'Gap' 열을 추가하시오.
$$ Gap = Close/Open - 1.0$$

```
### 답안을 작성하시오.
goog['Gap']=goog['Close']/goog['Open']-1.0
aapl['Gap']=aapl['Close']/aapl['Open']-1.0
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
goog.head()
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
aapl.tail()
```

### 문제2
두 변수 *goog*, *aapl* 에 대하여 아래 수식으로 정의되는 'Diff' 열을 추가하시오.
$$ {\rm Diff} = \frac{(당일 Close)}{(전일 Close)} ~~ - 1$$

- 예) 2010-01-05일 Diff값은 2010-01-04일 Close에 비해 2010-01-05일 Close가 변한 비율을 의미한다.
- nan이 발생해도 상관없다.

```
### 답안을 작성하시오.
goog['Diff']=goog['Close'].pct_change()
aapl['Diff']=aapl['Close'].pct_change()
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
goog.head()
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
aapl.tail()
```

### 문제4
두 변수 *goog*, *aapl* 에 대하여 아래 수식으로 정의되는 'GapL'과 'DiffL' 열을 추가하시오.
$$ {\rm GapL} = [{\rm Gap} \times 100]$$
$$ {\rm DiffL} = [{\rm Diff} \times 100]$$

단, $[x]$는 $x$보다 크지 않은 최대 정수를 의미한다.

- 'GapL'과 'DiffL' 열은 정수형 배열이다.

```
### 답안을 작성하시오.
goog['GapL']=(goog['Gap']*100).astype(int)
goog['DiffL']=(goog['Diff']*100).astype(int)
aapl['GapL']=(aapl['Gap']*100).astype(int)
aapl['DiffL']=(aapl['Diff']*100).astype(int)
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
goog.head()
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
aapl.tail()
```

### 문제5
두 변수 *goog*, *aapl* 에 대하여 다음 조건으로 pivot table을 구하여 각각 *pivot_goog*, *pivot_aapl*에 저장하시오.


* index와 column은 'GapL'과 'DiffL'을 사용한다.
* 'Volume'을 대상으로 개수 정보를 저장한다.
* nan값이 많은 경우 (단, axis=1에 대해 thres=7 기준) 삭제한다.



```
### 답안을 작성하시오.
pivot_goog=pd.pivot_table(data=goog,values='Volume',index='GapL',columns='DiffL',aggfunc='count',dropna=False)
pivot_aapl=pd.pivot_table(data=aapl,values='Volume',index='GapL',columns='DiffL',aggfunc='count',dropna=False)

pivot_goog=pivot_goog.dropna(axis=1, thresh=7)
pivot_aapl=pivot_aapl.dropna(axis=1, thresh=7)
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
pivot_goog
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
pivot_aapl
```

### 문제10

### 기타 (평가에 미반영)
변수 *two*에서 매월 마지막날의 'Close'가 그 달 첫째날의 'Close' 값에 비해 변화한 비율(아래 수식 참고)을 *mon_diff* 변수에 저장하시오.

$$ \frac{해당월의 마지막 날의 Close 값}{해당월의 첫째 날의 Close 값} - 1$$

- 인덱스는 해당월의 마지막 날짜이다.

힌트) DataFrame의 resample('BME')를 활용한다.

```
### 답안을 작성하시오.
close=two['Close']

first=close.resample('M').first()
last=close.resample('M').last()
mon=(last/first)-1
```

```
### 다음과 같이 출력되어야 한다. (수정하지 마시오.)
mon
```
